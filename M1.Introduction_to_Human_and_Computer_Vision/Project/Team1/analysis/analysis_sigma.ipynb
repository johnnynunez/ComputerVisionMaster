```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('metrics.csv')
```

```python
df
```

```python
# plot unamed: 0 in x and sigma in y in scatter plot with all values in x axis
fig = plt.figure(figsize=(10,8))
plt.scatter(df['Unnamed: 0'], df['sigma'])
plt.xticks(df['Unnamed: 0'])
plt.yticks(np.arange(0, 15, 1))
```

```python
fig = plt.figure(figsize=(10,8))
plt.scatter(df['Unnamed: 0'], df['sigma_denoised'])
plt.xticks(df['Unnamed: 0'])
plt.yticks(np.arange(0, 4, 0.5))
```
